```python
import numpy as np
import cv2
image = cv2.imread("/content/drive/MyDrive/Vision/flint1.jpg")
print(image.shape)

mask = np.zeros(image.shape[:2],dtype="uint8")

(cX, cY) = (image.shape[1] //2, image.shape[0] // 2)

cv2.rectangle(mask,(cX-75,cY-75),(cX+75,cY+75),225,-1)
cv2.imwrite("mask1.png",mask)
masked = cv2.bitwise_and(image,image,mask=mask)
cv2.imwrite("/content/drive/MyDrive/Vision/masked1.png",masked)

mask = np.zeros(image.shape[:2],dtype="uint8")
cv2.circle(mask,(cX,cY),100,225,-1)
cv2.imwrite("mask2.png",mask)
masked = cv2.bitwise_and(image,image,mask=mask)
cv2.imwrite("/content/drive/MyDrive/Vision/masked2.png",masked)
```
